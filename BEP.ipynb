```python
# BEP - Break Even Point ( Margin of Safety)
# Ofshore is relevant only for larger Demands. For small local demands is the local factory enough.
```

```python
import matplotlib.pyplot as plt
```

```python
# 1. Expensive Product, e.g. Blazer price 100 EUR
# Solution: e.g. local factory - we must sell min. 500 units to get a profit
```

```python
# Sample data
fixed_costs = 30000         # Monthly fixed costs in EUR
selling_price = 100         # Unit selling price
variable_cost = 40          # Variable cost per unit
```

```python
# Break-even calculation
pe_units = fixed_costs / (selling_price - variable_cost)
pe_revenue = pe_units * selling_price

# Output results
print(f"Break-even Point: {pe_units:.0f} units") # how many units do we need to sell to receive some profit
print(f"Required Revenue: EUR{pe_revenue:.2f}") # how much revenue we must make to make a profit
```

```python
# Generate chart
quantities = list(range(0, 1001, 50))
revenues = [q * selling_price for q in quantities]
total_costs = [fixed_costs + (q * variable_cost) for q in quantities]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(quantities, revenues, label='Revenue', color='grey')
plt.plot(quantities, total_costs, label='Total Costs', color='tan')
plt.axvline(pe_units, color='black', linestyle='--', label='Break-even Point')
plt.title('Break-even Point: Revenue vs Total Costs')
plt.xlabel('Units Sold')
plt.ylabel('Amount (EUR)')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```
